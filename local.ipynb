```python

db_path = '/Users/tanakatooryu/dsp2final/tanakatorudsp/local.db'
conn = sqlite3.connect(db_path)


cursor = conn.cursor()

query = "SELECT reason_for_not_choosing_win FROM mac_class"
cursor.execute(query)
data = cursor.fetchall()

conn.close()

reasons = [item[0] for item in data]

reason_counts = dict()
for reason in reasons:
    reason_counts[reason] = reason_counts.get(reason, 0) + 1

labels = list(reason_counts.keys())
sizes = list(reason_counts.values())


plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Reasons for not choosing win')
plt.show()
```
